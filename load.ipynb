```python
import os

import numpy as np
import tensorflow as tf

from PIL import Image
```

```python
bim = open('./test.jpg', 'rb').read()
```

```python
m = tf.saved_model.load('./preprocess_saved/')
```

```python
img_normed = m(bim).numpy()
r, g, b = img_normed[0]
r = r * 0.229 + 0.485
g = g * 0.224 + 0.456
b = b * 0.225 + 0.406

Image.fromarray(np.transpose((np.stack([r, g, b]) * 255).astype('uint8'), (1, 2, 0)))
```
